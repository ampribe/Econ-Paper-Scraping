```python
import regex as re
import pandas as pd
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
```

```python
# Enter link to 'issues' page of journal -- Example: https://www.aeaweb.org/journals/mic/issues
issue_index_link = 'ISSUE_INDEX_LINK'
index = BeautifulSoup(requests.get(issue_index_link).content)
```

```python
# Compile links to all issues from 2010-Present -- Stored as list of tuples (issue_link, issue)
all_links = index.find('section', class_='journal-preview-group').find_all('a')

# Regex matches issues from 2010-Present -- To include all issues, remove the regex
issue_links = [('https://aeaweb.org' + link['href'], link.text) for link in all_links if bool(re.match('.+?\d\d[12]\d', link.text))]
```

```python
# collect links to all articles -- Stored as list of tuples (article_link, issue)

driver = webdriver.Chrome()
all_article_links = []
tot = len(issue_links)
for i, (link, issue) in enumerate(issue_links):
    print(f'{i+1}/{tot}')
    driver.get(link)
    issue_text = BeautifulSoup(driver.page_source)
    articles_raw = issue_text.find_all('article', class_=re.compile('journal-article.+'))
    article_links = ([('https://aeaweb.org' + article.find('a', {'href': True})['href'], issue) 
                      for article in articles_raw if article.find('a', {'href': True})])
    all_article_links += article_links
driver.close()
```

```python
# Visits all article links to collect download links
# download links stored as list of tuples (article_link, download_link, title, issue)

driver = webdriver.Chrome()
tot = len(all_article_links)
all_download_links = []
for i, (link, issue) in enumerate(all_article_links):
    print(f'{i+1}/{tot}')
    driver.get(link)
    article_text = BeautifulSoup(driver.page_source)
    title = article_text.find('h1', class_='title').text
    download_link = 'https://aeaweb.org' + article_text.find('section', class_='download').find('a', class_='button')['href']
    if title != 'Front Matter':
        all_download_links.append((link, download_link, title, issue))
driver.close()
```

```python
links_df = pd.DataFrame(all_download_links, columns=['article_link', 'download_link', 'title', 'issue'])
```

```python
# Saves download links and metadata as csv -- enter csv name, example: 'aej_micro_download_links.csv'
links_df.to_csv('ENTER_FILENAME.csv', index=False)
```
